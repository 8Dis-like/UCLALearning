**Problem**
Westwood manufacturing company produces two types of eco-friendly packaging products from recycled materials: Boxes and Envelopes. The production process generates waste material, and the company is looking to minimize this waste while meeting the demand for both products. The company operates with certain limitations on the availability of raw materials and production capacity.

**SPECS**
Raw Material Availability: 1000 kg of recycled pulp per day.

**Product Specifications**
Box: Requires 3 kg of pulp to produce one unit. Produces 0.5 kg of waste per unit.

Envelope: Requires 1.5 kg of pulp to produce one unit. Produces 0.25 kg of waste per unit.

**Demand:** Daily demand is 200 units for Boxes and 300 units for Envelopes.

**Objective:** Minimize total waste from production.

**Decision Variables**
Let x represent the number of Boxes produced per day.
Let y represent the number of Envelopes produced per day.

**Objective Function**
Minimize the total waste produced:

Minimize W = 0.5x + 0.25y

**Constraints**
Raw material usage: 3x + 1.5y ≤ 1000

Demand satisfaction: x ≥ 200, y ≥ 300

Non-negativity: x , y ≥ 0

```python
from pulp import LpMaximize, LpProblem, LpStatus, LpVariable, LpMinimize
```

```python
# Set up minimization model
waste_model = LpProblem("Minimize_Waste", LpMinimize)
```

```python
# Define the decision variables
x = LpVariable("Boxes", lowBound=0, cat='Continuous') 
y = LpVariable("Envelopes", lowBound=0, cat='Continuous')
```

```python
# Define Objective Function
obj_func = 0.5 * x + 0.25 * y
# Add Objective function to the model
waste_model += obj_func
```

```python
# Add constraints to the model
waste_model += (3 * x + 1.5 * y <= 1000, "RawMaterialLimit")
waste_model += (x >= 200, "BoxDemand")
waste_model += (y >= 300, "EnvelopeDemand")
# x >=0
# y ≥ 0
```

```python
# review model setup
waste_model
```

```python
status = waste_model.solve()
print ("status returned code :", status)
print ({0: 'Not Solved', 1: 'Optimal', -1: 'Infeasible', -2: 'Unbounded', -3: 'Undefined'})
```

```python
print("Solution is", LpStatus[waste_model.status])
print("objective :", waste_model.objective.value())

for var in waste_model.variables():
    print(var.name, ":", var.value())
for name, constraint in waste_model.constraints.items():
    print(name, " : ", constraint.value())
```

Management Actions:

1) Re-evaluate Constraints: Ensure that the constraints accurately reflect the real-world scenario. This might include adjusting the raw material limit or re-assessing the waste generated per unit of production.

2) Flexibility in Demand: Look at the possibility of flexibility in demand constraints (e.g., allowing for a range rather than fixed values)to provide the model with the necessary leeway to find a feasible solution.
