## Spatial batch normalization

In fully connected networks, we performed batch normalization on the activations.  To do something equivalent on CNNs, we modify batch normalization slightly.  

Normally batch-normalization accepts inputs of shape `(N, D)` and produces outputs of shape `(N, D)`, where we normalize across the minibatch dimension `N`. For data coming from convolutional layers, batch normalization accepts inputs of shape `(N, C, H, W)` and produces outputs of shape `(N, C, H, W)` where the `N` dimension gives the minibatch size and the `(H, W)` dimensions give the spatial size of the feature map.

How do we calculate the spatial averages?  First, notice that for the `C` feature maps we have (i.e., the layer has `C` filters) that each of these ought to have its own batch norm statistics, since each feature map may be picking out very different features in the images.  However, within a feature map, we may assume that across all inputs and across all locations in the feature map, there ought to be relatively similar first and second order statistics.  Hence, one way to think of spatial batch-normalization is to reshape the `(N, C, H, W)` array as an `(N*H*W, C)` array and perform batch normalization on this array.

## Question

In our architecture, we apply batch normalization *before* ReLU. What would change if we applied it *after* ReLU instead? Which ordering do you think is better, and why?

## Answer
What changes if applied after ReLU: ReLU outputs are non-negative (including many zeros). If you apply BN after ReLU, the normalization step (subtracting the mean) will shift the data, making roughly half of the values negative again. This destroys the sparsity (the zero values) introduced by ReLU.
Statistics: BN assumes the input distribution is somewhat symmetric or Gaussian-like to estimate mean/variance effectively. ReLU outputs are not symmetric, which might make the statistics less stable or meaningful.

Which is better: BN before ReLU is generally considered better and is the standard practice.

Why:
 Stabilizing Inputs to Activation: The goal of BN is to normalize the inputs so they fall in a stable range. By placing BN before the activation, you ensure the values entering the ReLU are centered (usually zero-mean, unit variance). This ensures that roughly 50% of the neurons are active and 50% are inactive (rectified) at initialization, preventing the "dead ReLU" problem and allowing gradients to flow efficiently.
