## Business Problem: Facility Location Cost Minimization

A company is considering opening new facilities in three possible locations: A, B, and C.

Each location has a different fixed opening cost:

Facility A: $5

Facility B: $7

Facility C: $4

Due to operational requirements, the company must open at least one facility to maintain market presence and serve customers.

However, leadership wants to minimize total fixed opening costs while satisfying this requirement.

Decision Question

Which facility (or facilities) should the company open in order to:

Meet the requirement of opening at least one location, and

Minimize total opening cost?

```python
from pulp import LpMinimize, LpProblem, LpVariable, lpSum
```

```python
# Define the problem as a minimization problem - minimizes the total opening cost
model = LpProblem("Facility_Location", LpMinimize)
```

```python
# List of facilities
facilities = ["A", "B", "C"]

# Define binary decision variables for each facility
x = {f: LpVariable(f"x_{f}", 0, 1, cat="Binary") for f in facilities}
```

```python
# Define costs associated with opening each facility
costs = {"A": 5, "B": 7, "C": 4}

# Objective function: Minimize the total opening cost
model += lpSum(costs[f] * x[f] for f in facilities)
```

```python
# Constraint: At least one facility must be opened
model += lpSum(x[f] for f in facilities) >= 1
```

```python
model
```

```python
# Solve the optimization problem
model.solve()
```

```python
# Print the results
for f in facilities:
    print(f"Facility {f} is opened: {x[f].value()}")
```

Since we are minimizing the cost and <strong> at least one facility must be opened </strong>, the solver selects the facility with the lowest cost.
<p>Facility C has the lowest cost (4), so it is selected (x_C = 1).
<p>Facilities A and B have higher costs and are not opened (x_A = 0, x_B = 0).
