```python
# Import PuLP
from pulp import LpProblem, LpVariable, LpMinimize, value

# Step 1: Define the Problem
# Create the Linear Programming problem
model = LpProblem("Minimize_Shipping_Costs", LpMinimize)

# Step 2: Define the Decision Variables
# Explicitly declare the variables for the routes
x_seattle_la = LpVariable("x_seattle_la", lowBound=0)       # Seattle to LA
x_seattle_chicago = LpVariable("x_seattle_chicago", lowBound=0)  # Seattle to Chicago
x_dallas_chicago = LpVariable("x_dallas_chicago", lowBound=0)    # Dallas to Chicago
x_dallas_atlanta = LpVariable("x_dallas_atlanta", lowBound=0)    # Dallas to Atlanta
x_miami_ny = LpVariable("x_miami_ny", lowBound=0)           # Miami to NY
x_miami_atlanta = LpVariable("x_miami_atlanta", lowBound=0)       # Miami to Atlanta

# Step 3: Define the Objective Function
# Minimize total shipping cost
model += (
    100 * x_seattle_la +
    120 * x_seattle_chicago +
    110 * x_dallas_chicago +
    90 * x_dallas_atlanta +
    95 * x_miami_ny +
    85 * x_miami_atlanta
), "Total_Shipping_Cost"

# Step 4: Add Constraints
# Warehouse capacity constraints
model += x_seattle_la + x_seattle_chicago <= 400, "Seattle_Capacity"
model += x_dallas_chicago + x_dallas_atlanta <= 300, "Dallas_Capacity"
model += x_miami_ny + x_miami_atlanta <= 200, "Miami_Capacity"

# Supermarket demand constraints
model += x_seattle_la + x_dallas_chicago + x_miami_atlanta == 200, "LA_Demand"
model += x_seattle_chicago + x_dallas_chicago == 150, "Chicago_Demand"
model += x_dallas_atlanta + x_miami_atlanta == 250, "Atlanta_Demand"
model += x_miami_ny == 100, "NY_Demand"

# Step 5: Solve the Model
model.solve()

# Step 6: Print Results
print("Optimal Total Cost:", value(model.objective))
print("Optimal Shipping Quantities:")
print(f"Seattle to LA: {x_seattle_la.varValue}")
print(f"Seattle to Chicago: {x_seattle_chicago.varValue}")
print(f"Dallas to Chicago: {x_dallas_chicago.varValue}")
print(f"Dallas to Atlanta: {x_dallas_atlanta.varValue}")
print(f"Miami to NY: {x_miami_ny.varValue}")
print(f"Miami to Atlanta: {x_miami_atlanta.varValue}")
```
