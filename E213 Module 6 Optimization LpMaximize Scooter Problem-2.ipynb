```python
from pulp import LpMaximize, LpProblem, LpStatus, LpVariable, LpMinimize
```

```python
# Set up maximization model
scooter_model = LpProblem(name='Scooter', sense=LpMaximize)
```

```python
# Define the decision variables
R = LpVariable(name="R", lowBound=0, upBound = 5000, cat = 'Integer')
N = LpVariable(name="N", lowBound=0, upBound = 5000, cat='Integer')
```

```python
# Define Objective Function
obj_func = 150*R + 160*N
# Add Objective function to the model
scooter_model += obj_func
```

```python
# Add constraints to the model
scooter_model += (4*R + 5*N <= 5610, "Frame Manufacturing")
scooter_model += (1.5*R + 2.0*N <= 2200, "Wheels and Deck Assembly")
scooter_model += (1.0*R + 0.8*N <= 1200, "QA and Packaging")
# R, N = integer
# R, N >= 0
```

```python
# review model setup
scooter_model
```

```python
status = scooter_model.solve()
print ("status returned code :", status)
print ({0: 'Not Solved', 1: 'Optimal', -1: 'Infeasible', -2: 'Unbounded', -3: 'Undefined'})
```

```python
print("Solution is", LpStatus[scooter_model.status])
print("objective :", scooter_model.objective.value())

for var in scooter_model.variables():
    print(var.name, ":", var.value())
for name, constraint in scooter_model.constraints.items():
    print(name, " : ", constraint.value())
```
