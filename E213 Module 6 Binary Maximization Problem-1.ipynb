## Business Problem: Strategic Facility Expansion

A company is evaluating whether to open facilities in three potential locations: A, B, and C.

Each facility requires an upfront investment (fixed opening cost):

Facility A: 5

Facility B: 7

Facility C: 4

Management has decided that at least one facility must be opened to ensure market presence.

## Unlike a cost-minimization problem, this model is set up as a maximization problem, meaning leadership believes that opening facilities generates strategic value (for example: revenue potential, market share impact, or long-term positioning benefits) proportional to the investment amount.

## The company wants to:

Decide which facilities to open (open = 1, not open = 0)

Open at least one facility

Maximize total strategic value generated from opened facilities

```python
"""
LpMaximize: Indicates that we are solving a maximization problem.
LpProblem: Creates an optimization problem instance.
LpVariable: Defines decision variables.
lpSum: Used to sum multiple terms in the objective function.
"""
from pulp import LpMaximize, LpProblem, LpVariable, lpSum
```

```python
# Define the problem
# Creating the Optimization Model

model = LpProblem("Facility_Location", LpMaximize)
```

```python
# Defining Decision Variables
# Define binary variables for facility locations

facilities = ["A", "B", "C"]

x = {f: LpVariable(f"x_{f}", 0, 1, cat="Binary") for f in facilities}
```

```python
# Objective function: Minimize cost 

costs = {"A": 5, "B": 7, "C": 4}

# maximizing the sum of the costs associated with opening the facilities

model += lpSum(costs[f] * x[f] for f in facilities)   

# for a maximization problem, the objective is try to open all facilities to achieve the highest possible value
```

```python
# Constraints: At least one facility must be opened

model += lpSum(x[f] for f in facilities) >= 1
```

```python
# Solve the model
model.solve()
```

```python
# Print results
for f in facilities:
    print(f"Facility {f} is opened: {x[f].value()}")
```

## The objective is to maximize the sum of facility values

There is no budget constraint limiting total investment

And only the constraint “open at least one facility” is imposed

The optimal solution is to open all facilities, since doing so increases the objective value.

## This model is intentionally incomplete from a managerial standpoint because:

There is no capital budget constraint.

There is no diminishing return or demand constraint.

There is no operational capacity limit.

## This is to illustrate:

Why constraints matter in optimization.

Why maximizing revenue alone is not realistic.

Why business models require trade-offs (budget, risk, operational feasibility).
